# 기능의 편의

```python
y=[2,5,8,9]

import matplotlib.pyplot as plt
plt.plot(y,marker='*')
```

```python
plt.xlim([1,3])
plt.ylim([1,10])
plt.plot(y,marker='*')
```

```python
plt.axis([1,3,1,10])
plt.plot(y,marker='*')
```

```python
x=[1,2,3]
y=[4,5,6]
y1=[2,4,6]
y2=[7,8,9]
y3=[10,30,40]
plt.plot(x,y,'-')
plt.plot(x,y1,'--')
plt.plot(x,y2,':')
plt.plot(x,y3,'-.')
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
plt.plot(x,y,'r*-')
```

```python
#plt.fill #영역 표시
#plt.fill_between() #수평 영역
#plt.fill_betweenx #수직 영역

x=[1,2,3,4]
y=[4,5,6,7]
plt.plot(x,y)
plt.fill_between(x[1:3],y[1:3],alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]

plt.plot(x,y)
plt.plot(x,y1)
plt.fill_between(x[1:3],y[1:3],y1[1:3],alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]

plt.plot(x,y)
plt.plot(x,y1)
plt.fill([2.0,2.0,3.5,3.5],[2.0,6.0,6.0,3.0],alpha=0.5)
```

```python
import numpy as np
x=np.linspace(-10,10,100)
y=x**3
plt.plot(x,y)
plt.xscale('symlog')
```

```python
x=np.linspace(0,10,100)
y=x**3
plt.plot(x,y)
plt.yscale('log')
```

```python
x=np.linspace(0,10,100)
y=np.exp(x)
plt.plot(x,y)
plt.yscale('log')
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]

plt.plot(x,y)
plt.plot(x,y1)
plt.grid(True,c='b',ls='--',alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
plt.plot(x,y)
plt.xticks([1,2,3,4],labels=['data1','data2','data3','data4'])
plt.yticks([4,5,6,7],labels=['a','b','c','d'])
```

```python
x=np.arange(0,4,0.5)
plt.plot(x,x+1)
plt.plot(x,x**2-4)
plt.plot(x,-2*x+3)
#plt.axhline(시작,시작점%,끝점%) #축수평
#plt.axvline #축수직
#plt.hlines(시작,시작점,끝점) #점수평
#plt.vlines #점수직
plt.axhline(2,0.5,1,c='r',ls='--')
plt.hlines(0.6,1.5,2.5,colors='b',ls=':')
plt.axvline(1.7,0.1,0.8,c='g',ls='--')
plt.vlines(0.3,2,4,colors='k',ls=':')
```
