```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame

data_l=[1,2,3,4,5]
e_obj=Series(data_l)
e_obj
```

```python
idx_l=[1,2,3,4,1]
e_obj=Series(data=data_l, index=idx_l)
e_obj
```

```python
idx_l=['a','b','c','d','e']
e_obj=Series(data=data_l, index=idx_l)
e_obj
```

```python
e_obj=Series(data=[data_l,data_l,data_l,data_l,1], index=idx_l)
e_obj
```

```python
e_obj.name="data_s"
e_obj.index.name="id"
e_obj
```

```python
dic={"a":1, "b":2, "c":3, "d":4, "e":5}
e_obj=Series(dic,dtype=float, name="data_s")
e_obj
```

```python
idx_d=['a','b','c','d','e','f','h']
dic={"a":1, "b":2, "c":3, "d":4, "e":5, "g":6}
e_obj=Series(dic, index=idx_d)
e_obj
```
