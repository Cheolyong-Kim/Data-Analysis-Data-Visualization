```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
import pandas as pd
```

```python
df = pd.read_csv('ex2_data.csv', encoding= 'cp949')
df
```

```python
text = df['내용']
text
```

```python
ok_t = Okt()
data1 = []

for i in text:
    t = ok_t.nouns(i)
    data1.extend(t)
data1
```

```python
data2 = [i for i in data1 if len(i) > 1]
data2
```

```python
data3 = Counter(data2)
data3
```

```python
data4 = data3.most_common(100)
data4
```

```python
data5 = dict(data4)
data5
```

```python
wc = WordCloud(font_path= 'C:\Windows\Fonts\malgun.ttf', background_color= 'white')
g_data = wc.generate_from_frequencies(data5)
plt.figure(figsize= (10,10))
plt.axis('off')
plt.imshow(g_data)
plt.show()
plt.savefig('image_q2.png')
```
